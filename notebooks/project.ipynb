```python
import re

# Lista de archivos
Filenames = [
    '../data/baby1990.html', 
    '../data/baby1992.html',
    '../data/baby1994.html',
    '../data/baby1996.html',
    '../data/baby1998.html',
    '../data/baby2000.html',
    '../data/baby2002.html',
    '../data/baby2004.html',
    '../data/baby2006.html',
    '../data/baby2008.html',
]

# Crear un diccionario
resultados = {}

    # 1. Leer el contenido del archivo HTML
for filename in Filenames:
    with open(filename, 'r', encoding='utf-8') as file:
        html_content = file.read()

    #2. Encontrar el año utilizando una expresión regular
    year_match = re.search(r"Popularity in (\d{4})", html_content)
    if year_match:
        year = year_match.group(1)

    # 3.Extraer los nombres de niños y niñas con el correspondiente número de ranking
    names_data = re.findall(r"<tr align=\"right\"><td>(\d+)</td><td>(\w+)</td><td>(\w+)</td>", html_content)

    #4. Agregar los nombres al diccionario de resultados
    for rank, boy_name, girl_name in names_data:
        resultados[boy_name] = f"Año: {year}, Posición: {rank}"
        resultados[girl_name] = f"Año: {year}, Posición: {rank}"

#5. Crear una lista de nombres ordenados alfabéticamente
sorted_names = sorted(resultados.keys())

#6. Imprimir la lista ordenada alfabéticamente con el año y su posicion
for name in sorted_names:
    print(f"{name}: {resultados[name]}")
```
